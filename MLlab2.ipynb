```python
12
import pandas as pd

def find_min_max(df, column_name):
    if column_name in df.columns:
        min_value = df[column_name].min()
        max_value = df[column_name].max()
        return {'Min Value': min_value, 'Max Value': max_value}
```

```python
result = find_min_max(df, 'Price in thousands')
print(result)
```

```python
13
hp_result = find_min_max(df, 'Horsepower')
print("Horsepower:", hp_result)

# Find min and max for Length
length_result = find_min_max(df, 'Length')
print("Length:", length_result)

# Find min and max for Fuel Efficiency
fuel_efficiency_result = find_min_max(df, 'Fuel Efficiency')
print("Fuel Efficiency:", fuel_efficiency_result)
```

```python
from sklearn.preprocessing import OneHotEncoder

# Selecting categorical column
categorical_cols = ['Vehicle type']

# Initialize OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)  # Use False for DataFrame output

# Fit and transform the column
encoded_array = encoder.fit_transform(df[categorical_cols])

# Convert to DataFrame with correct column names
encoded_df = pd.DataFrame(encoded_array, columns=encoder.get_feature_names_out(categorical_cols))

# Merge back with original dataset
df_final = pd.concat([df, encoded_df], axis=1).drop(columns=categorical_cols)

print(df_final.head())
```
